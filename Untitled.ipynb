```python
from pykrx import stock

ticker_lst = stock.get_market_ticker_list(date='20250707')

for idx, ticker in enumerate(ticker_lst):
    ticker_name = stock.get_market_ticker_name(ticker)
    print(ticker_name)
    if idx >= 5:
        break
```

```python
rslt = stock.get_market_ohlcv('20250707')
print(rslt)
```

```python
rslt_fund = stock.get_market_fundamental('20250707')
print(rslt_fund)
```

```python
rslt_tot_df = rslt.join(rslt_fund,how='inner')
print(rslt_tot_df)
```

```python
rslt_tot_df = rslt_tot_df.sort_values(by=['등락률'],ascending=False)
rslt_tot_df.reset_index(inplace=True)
print(rslt_tot_df)
```

```python
for idx, row in rslt_tot_df.iterrows():
    ticker_name = stock.get_market_ticker_name(row['티커'])
    fluc_rate = row['등락률']
    print(ticker_name)
    print(fluc_rate)

    if idx >= 4:
        break
```
